```python
Array Generation Function
```

```python
arr = np.arange(1,21,3) # arange(start,end,step)
arr
```

```python
arr = np.zeros(3)
arr
```

```python
arr = np.zeros((3,6)) # zeros(row,col)
arr
```

```python
arr = np.ones(4)
arr
```

```python
arr = np.ones((4,10))
arr
```

```python
arr = np.linspace(1,5,20) # linear Space between start and end linspace(start,end,space_no)
arr
```
