# Deep and Shallow Copy

```python
# Shallow copy:dosen't change the original array creates new memory
# Deep copy:changes the content
# slice portion gets new memory
# but if copied content make of an array then it dosen't create new memory
```

```python
# creating vector array
a4 = np.arange(1,21)
a4
```

```python
slice = a4[:5]
slice = slice*10
slice
```

```python
a4 # Shallow copy:dosen't change the original array creates new memory
```

```python
a4[3:6]*10
```

```python
a4
```

```python
# Deep copy
b1 = a4
b1
```

```python
b1[0] = 33 # changing the content of b1 will also change content in a4
b1
```

```python
a4
```

```python
# but if copy createdof a4 then b1 changes will not get seen in a4
b1 = a4.copy()
b1[4] = 43
b1
```

```python
a4 # here changes are not in a4 but b1 only
```
